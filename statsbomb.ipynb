### Free matches availables Statsbomb 

```python
messages = []
for index, row in df_competitions.iterrows():
    try:
        competition = row['competition_id']
        season = row['season_id']
        country_name = row['country_name']
        competition_name = row['competition_name']
        season_name = row['season_name']
        number_matches = len(sb.matches(competition_id=competition, season_id=season))
        message = f'{country_name} {competition_name} {season_name} has {number_matches} available'
        messages.append(message)
    except:
        pass
```

```python
for message in messages:
    print(message)
```
